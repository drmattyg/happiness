```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
h = 128
w = 128
def happiness(b):
    x = np.arange(0, w - 1, 1)
    def f(x):
        return (b*h/2)*np.sin(x*np.pi/w)
    plt.plot(x, f(x), '-')
    
```

```python
plt.subplot(3, 3, 1)
happiness(-1)
plt.subplot(3, 3, 2)
happiness(0)
plt.subplot(3, 3, 3)
happiness(1)
```

```python
happiness(0)
```

```python
happiness(1)
```

```python
def happy_circle(r):
    x = np.arange(0, w, 1)
    d = 0.5*np.sqrt(4*r**2 - (w**2))
    y0 = h/2 + d
    x0 = w/2
    y = -np.sqrt(r**2 - ((x - x0) **2)) + y0
    plt.xlim(0, w)
    plt.ylim(0, h)
    plt.plot(x, y)
```

```python
happy_circle(w/2)
```

```python
happy_circle(w)
```
