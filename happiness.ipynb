```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
h = 128
w = 128
def happiness(b):
    x = np.arange(0, w - 1, 1)
    def f(x):
        return (b*h/2)*np.sin(x*np.pi/w)
    plt.plot(x, f(x), '-')
    
```

```python
plt.subplot(3, 3, 1)
happiness(-1)
plt.subplot(3, 3, 2)
happiness(0)
plt.subplot(3, 3, 3)
happiness(1)
```

```python
happiness(0)
```

```python
happiness(1)
```

```python
def happy_circle(v):
    if v == 0:
        v += 0.0001
    r = (w/2)/np.abs(v)
    x = np.arange(0, w, 1)
    d = 0.5*np.sqrt(4*r**2 - (w**2))
    y0 = h/2 + np.sign(v)*d
    x0 = w/2
    y = -np.sign(v)*np.sqrt(r**2 - ((x - x0) **2)) + y0
    plt.xlim(0, w-1)
    plt.ylim(-10, h + 10)
    plt.plot(x, y)
```

```python
for v in np.arange(-1.0, 1.2, 0.2):
    happy_circle(v)
```

```python
happy_circle(1)
```

```python
happy_circle(-1)
```

```python
happy_circle(0)
```
