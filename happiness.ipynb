```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
h = 128
w = 128
def happiness(b):
    x = np.arange(0, w - 1, 1)
    def f(x):
        return (b*h/2)*np.sin(x*np.pi/w)
    plt.plot(x, f(x), '-')
    
```

```python
plt.subplot(3, 3, 1)
happiness(-1)
plt.subplot(3, 3, 2)
happiness(0)
plt.subplot(3, 3, 3)
happiness(1)
```

```python
happiness(0)
```

```python
happiness(1)
```

```python
def happy_circle(r):
    x = np.arange(0, w - 1, 1)
    y0 = np.sqrt(r**2 - (w**2)/4)
    y = np.sqrt(r - x**2) - y0
    plt.plot(x, y)
```

```python
happy_circle(h/2)
```

```python
happy_circle(h*10)
```
